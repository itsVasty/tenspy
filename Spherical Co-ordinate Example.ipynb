```python
from sympy import *
import numpy as np
```

```python
r, theta, phi = symbols('r theta phi')
init_printing(use_unicode=True)
```

## Initialising Problem Space

```python
x = r*sin(phi)*cos(theta)
y = r*sin(phi)*sin(theta)
z = r*cos(phi)

C =  [x, y, z]
R = [r, theta, phi]
```

## Finding Basis Vectors

```python
g = []
for i in R:
    g.append([diff(j,i) for j in C])
Matrix(g)
```

## Finding Metric Tensor (G)

```python
G = []
for a in g:
    g_ = []
    for b in g:
        g_.append(trigsimp(np.array(a)@np.array(b)))
    G.append(g_)
Matrix(G)
```

## Finding Christoffel Symbols

#### Finding Inverse Metric

```python
G_ = np.array(Matrix(G).inv())
Matrix(G_)
```

#### Defining Christoffel Function

```python
Gamma = lambda i,j,k: sum([0.5*G_[i][s]*(diff(G[s][j], R[k]) + diff(G[k][s], R[j]) - diff(G[j][k], R[s])) for s in range(len(R))])
```

##### _Example 1_: 
Find $\Gamma^{\theta}_{{\phi}{\phi}}\$$ and
$\Gamma^{r}_{{\theta}{\theta}}\$

i) $\Gamma^{\theta}_{{\phi}{\phi}}\$

```python
Gamma(1,2,2)
```

i) $\Gamma^{r}_{{\theta}{\theta}}\$

```python
Gamma(0,1,1)
```

##### _Example 2_: 
Find $\Gamma^{r}_{{\phi}{\phi}}\$$ and
$\Gamma^{\phi}_{{\theta}{\phi}}\$

i) $\Gamma^{r}_{{\phi}{\phi}}\$

```python
Gamma(0,2,2)
```

ii) $\Gamma^{\phi}_{{\theta}{\phi}}\$

```python
Gamma(2,1,2)
```

#### Christoffel Symbol Table

```python
def Christoffel_Table(n):
    T_ = []
    for j, _ in enumerate(R):
        t_ = []
        for k, _ in enumerate(R):
            t_.append(Gamma(n,j,k))
        T_.append(t_)
    return T_
```

1. Table for $\Gamma^{r}\$

```python
T_r = Christoffel_Table(0)
Matrix(T_r)
```

2. Table for $\Gamma^{\theta}\$

```python
T_t = Christoffel_Table(1)
Matrix(T_t)
```

3. Table for $\Gamma^{\phi}\$

```python
T_p = Christoffel_Table(2)
Matrix(T_p)
```

#### Riemann-Christoffel Tensor

```python
Riemm = lambda n,i,j,k: diff(Gamma(n,i,k), R[j]) - diff(Gamma(n,i,j), R[k]) + sum([Gamma(t,i,k)*Gamma(n,t,j) - Gamma(t,i,j)*Gamma(n,t,k) for t,_ in enumerate(R)])
```

##### _Example 3_: 
Find $R^{\theta}_{{\phi}{\phi}{\theta}}\$

```python
Riemm(0,1,1,0)
```

#### Riemann-Christoffel Tables

```python
def Riemman_Table(n,i):
    T_ = []
    for j,_ in enumerate(R):
        t_ = []
        for k,_ in enumerate(R):
            t_.append(Riemm(n,i,j,k))
        T_.append(t_)
    return T_
```

```python
Matrix(Riemman_Table(0,0)), Matrix(Riemman_Table(0,1)), Matrix(Riemman_Table(0,2))
```

```python
Matrix(Riemman_Table(1,0)), Matrix(Riemman_Table(1,1)), Matrix(Riemman_Table(1,2))
```
